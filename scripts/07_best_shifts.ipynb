```python
import pandas as pd
import glob
import statistics
import numpy as np
```

```python
path = '../data/best_shifts/'
```

```python
csvs = sorted(glob.glob(path+"*"))
df = pd.DataFrame(columns=['Framework', 'Subject', 'Exercise', 'View', 'best_shift'])
for csv in csvs:
    df_framework = pd.read_csv(csv, index_col=0)
    df_framework['Framework'] = csv.rsplit('_', 1)[-1].split('.', 1)[0]
    df = pd.concat([df, df_framework])
```

```python
result_df = df.groupby(['Subject', 'Exercise', 'View'], as_index=False)['best_shift'].agg(list)
result_df['best_shift_unique_value'] = result_df['best_shift'].apply(lambda x: {val: x.count(val) for val in set(x)})
result_df['best_shift_median'] = result_df['best_shift'].apply(lambda x: statistics.median(x) if not np.isnan(statistics.median(x)) else np.nan)
result_df
```

```python
result_df.to_csv('../data/best_shifts/best_shifts.csv', index=False)
```
